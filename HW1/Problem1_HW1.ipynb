```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
plt.style.use('seaborn-white')

# To clean up our code we encapuslate the graphing into a class
# n is our sample size
# we pass our functions as parameters
class PlotGraph:
  def plot(self, function, n=40, cmapOpt='viridis', cMap2d='black'):
    # Init plotting constructors and crunch some calculations
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    x = np.linspace(-n, n)
    y = np.linspace(-n, n)
    X, Y = np.meshgrid(x,y)
    Z = function(X, Y)

    # Plot our 3D and 2D graph
    ax.contour3D(X, Y, Z, 50, cmap=cmapOpt)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.show()
    plt.contour(X,Y,Z, cmap=cMap2d)
    plt.show()
    plt.contourf(X,Y,Z, cmap=cMap2d)
    plt.colorbar()
    plt.show()
    
# All the functions needed to perform calculations
# Matyas function
# https://en.wikipedia.org/wiki/Test_functions_for_optimization
def localMin(x, y):
  return 0.26*(x**2 + y**2) -0.48*x*y

def localMax(x, y):
  return (2*np.cos(x) + 2*np.sin(y))

# Himmelblau's function
# https://en.wikipedia.org/wiki/Himmelblau%27s_function
def noMinNoMax(x, y):
  return (x**2 + y -11)**2 + (x + y**2 - 7)**2
```

```
# Init our plot class
graphy = PlotGraph()

# Local Min
print("===================== Local Min =====================")
graphy.plot(localMin, cMap2d='Blues')
```

```
# Local Max
print("===================== Local Max =====================")
graphy.plot(localMax, cmapOpt='inferno', cMap2d='autumn')
```

```
# No local minimum or maximum
print("===================== No Min or Max =====================")
graphy.plot(noMinNoMax, cmapOpt='cubehelix', cMap2d='cubehelix')
```
